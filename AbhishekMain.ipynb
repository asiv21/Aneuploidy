```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

#%config InlineBackend.figure_format = 'svg'
```

```python
B1_IO = pd.read_csv('Data/Batch_1/Interval_order.csv')
B1_Depth = pd.read_csv('Data/Batch_1/Depth.csv')

B2_IO = pd.read_csv('Data/Batch_2/Interval_order.csv')
B2_Depth = pd.read_csv('Data/Batch_2/Depth.csv')

LenSeg1 = B1_IO['End']-B1_IO['Start']
LenSeg2 = B2_IO['End']-B2_IO['Start']

heads = list(B1_Depth.head())
B1_Depth_wt = B1_Depth.copy()
B2_Depth_wt = B2_Depth.copy()
for h in heads:
    B1_Depth_wt[h] = B1_Depth[h]/LenSeg1
    B2_Depth_wt[h] = B2_Depth[h]/LenSeg2
```

```python
bounds1=[] 
for i in range(1,25): 
    bounds1.append((B1_IO.loc[B1_IO['Chrom'] == i].iloc[0].name,B1_IO.loc[B1_IO['Chrom'] == i].iloc[-1].name)) 
Depth1_chrom = np.zeros((24,24)) # Chromosome number (1-24) (on the rows) x sample number (1-24) (on the columns)

for i in range(24):
    Depth1_chrom[i,:] = np.sum(B1_Depth[bounds1[i][0]:bounds1[i][1]+1])

Depth1_chrom_wt = np.zeros((24,24)) # Chromosome number (1-24) (on the rows) x sample number (1-24) (on the columns)
for i in range(24):
    Depth1_chrom_wt[i,:] = np.sum(B1_Depth_wt[bounds1[i][0]:bounds1[i][1]+1])
```

```python
bounds2=[] 
for i in range(1,25): 
    bounds2.append((B2_IO.loc[B2_IO['Chrom'] == i].iloc[0].name,B2_IO.loc[B2_IO['Chrom'] == i].iloc[-1].name)) 
Depth2_chrom = np.zeros((24,24)) # Chromosome number (1-24) (on the rows) x sample number (1-24) (on the columns)
for i in range(24):
    Depth2_chrom[i,:] = np.sum(B2_Depth[bounds2[i][0]:bounds2[i][1]+1]) 

Depth2_chrom_wt = np.zeros((24,24))
for i in range(24):
    Depth2_chrom_wt[i,:] = np.sum(B2_Depth_wt[bounds2[i][0]:bounds2[i][1]+1])
```
